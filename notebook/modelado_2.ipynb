# Modelado de base de datos imaginaria, para probar la efectividad del modelo
Se importan las librerías necesarias para la lectura de datos (pandas) y para cargar el modelo previamente entrenado (pickle).

```python
import pandas as pd
import pickle
```

Se carga la base de datos imaginaria, que contiene los registros sobre los cuales se desea realizar la segunda predicción.

```python
df_pred = pd.read_csv("../data/dataset_2.csv")
```

Se carga el modelo entrenado desde el archivo .pkl para evitar realizar un nuevo entrenamiento.

```python
with open("../data/modelo.pkl", "rb") as f:
    modelo = pickle.load(f)
```

Se genera una nueva columna llamada Predicted_Session, en la cual se guardan las predicciones realizadas por el modelo.

```python
df_pred["Predicted_Session"] = modelo.predict(df_pred)
```

Se exporta el nuevo dataset que incluye la columna con las predicciones generadas por el modelo.

```python
df_pred.to_csv("../data/tattoo_predictions_modelado2_con_predicciones.csv", index=False)
```
