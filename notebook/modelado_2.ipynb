# Modelado de base de datos imaginaria, para probar la efectividad del modelo
se importa panda y pickle para llamar las bases de datos y el modelo

```python
import pandas as pd
import pickle
```

se lee la base de datos imaginaria

```python
df_pred = pd.read_csv("../data/dataset_2.csv")
```

se llama el modelo para no volver a entrenar los datos

```python
with open("../data/modelo.pkl", "rb") as f:
    modelo = pickle.load(f)
```

se escoge la variable a predecir que quedara ilustrada como una columna

```python
df_pred["Predicted_Session"] = modelo.predict(df_pred)
```

se genera la nueva base de datos con la columna de la predicci√≥n.

```python
df_pred.to_csv("../data/tattoo_predictions_modelado2_con_predicciones.csv", index=False)
```
